# Copy vs View
- Definition: Copy creates a new independent array, while View shares the same data with the original array

```python
arr = np.array([1,2,3])
x = arr.view() # View -> changes reflect
y = arr.copy() # Copy -> independent
```

# Linear Algebra
- Definition: Linear algebra operations like matrix multiplication, inverse, and determinant are handled by NumPy.

```python
# Dot product 
a = np.array([1,2])
b = np.array([3,4])
np.dot(a,b)
```

```python
# Matrix Multiplication
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
A @ B
```

```python
# Determinant
np.linalg.det(A)
```

```python
# Inverse
np.linalg.inv(A)
```
