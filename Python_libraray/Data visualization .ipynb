```python
import pandas as pd
import matplotlib.pyplot as plt

# Group total sales by region
region = df.groupby('Region')['TotalSales'].sum()

# Convert Series to DataFrame
region = pd.DataFrame(region)

# Reset index
region.reset_index(inplace=True)

# Plot graph
plt.figure()
plt.plot(region['Region'], region['TotalSales'])
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Total Sales for each Region')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = x ** 2

plt.figure(figsize=(6, 6))

positions = [1, 3, 5, 7, 9]

for pos in positions:
    plt.subplot(3, 3, pos)
    plt.plot(x, y, marker='*', linestyle='--')
    plt.title(f"y = x² (Plot {pos})")
    plt.xlabel("x values")
    plt.ylabel("y values")
    plt.grid(True)

plt.suptitle(" Subplot Representation of y = x²", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

# pie chart:

- It is used to represent data in the form of percentage.

# Syntax:

 - plt.pie(data,label=[],autopct='%0,1f%%,color=[],shadow=True,explode=())

```python

viewrs = [1500,5000,2000,1000,2500]
l = ['Drama','Action','Horror','Comedy','Romantic']
plt.title("Movie  Viewers Distribution", fontsize=14, fontweight='bold')
plt.pie(viewrs,labels=l)
plt.pie(viewrs,labels=l,colors=['pink','gray','red','orange','yellow'],autopct='%0.2f%%' ,shadow=True) 

plt.show()
```

```python

viwers = [1500,5000,2000,1000,2500]
l = ['Drama','Action','Horror','Comedy','Danger']
plt.title("Movie  Viewers Distribution", fontsize=14, fontweight='bold')

plt.pie(viwers,labels=l,colors=['green','red','brown','yellow','orange'],autopct='%0.2f%%',shadow=True,explode=(0.1,0.0,0.1,0.1,0.1))
plt.show()

```
