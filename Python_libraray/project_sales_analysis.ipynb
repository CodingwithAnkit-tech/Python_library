```python
# import all necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Practice/sales_data.csv",encoding='ISO-8859-1')
df
```

```python
df.shape
```

```python
df.info()
```

```python
# dropping unwanted from dataset

df.drop(['Status','unnamed1'],axis=1,inplace=True)
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.head()
```

```python
# Rename the column Marital_Status to Marital
df.rename(columns={"Marital_Status": "Marital"}, inplace=True)
df.head()
```

```python
# change the value of Martial col 0---> single,1 --->Married
df['Marital'] = df['Marital'].map({0:'single',1:'married'})
df.head()
```

```python
df['Gender'] = df['Gender'].map({'F':'Female','M':'Male'})
df.head()
```

```python
# visulaize the count of males and female based on gender column using seaborn

a = sns.countplot(x='Gender',data=df,hue='Gender')
for i in a.containers:
    a.bar_label(i)

# plt.figure(figsize=(6,4))
# sns.countplot(x='Gender', data=df)

# plt.title("Count of Male and Female")
# plt.show()
```

```python
# group by gender
a = df.groupby('Gender')['Amount'].sum()
a = pd.DataFrame(a)

# visulaize male and female amount
a = sns.barplot(x='Gender',y='Amount',data=a,hue='Gender')
for i in a.containers:
    a.bar_label(i)
```
